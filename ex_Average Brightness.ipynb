```python
import cv2
import helpers
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

%matplotlib inline
```

```python
image_dir_training = "day_night_images/training/"
image_dir_test = "day_night_images/test/"
```

```python
#standalize all training image
STANDALIZED_LIST = helpers.load_dataset(image_dir_training)
```

```python
#display the standalize image and their label
image_num = 0
selected_image = STANDALIZED_LIST[image_num][0]
selected_label = STANDALIZED_LIST[image_num][1]

#display image and the data about it
plt.imshow(selected_image)
print("shape: ", selected_image.shape)
print("Lable(1=Day, 0=night): ", selected_label)
```

```python
#feature extraction
#convert image to HSV color space and visualize the image

image_num = 0
test_im = STANDALIZED_LIST[image_num][0]
test_label = STANDALIZED_LIST[image_num][1]

#convert image to hsv
hsv = cv2.cvtColor(test_im, cv2.COLOR_RGB2HSV)

#print image label
print("Label: " +str(test_label))

#HSV channels
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

#display
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize = (20, 10))

ax1.set_title("standalized image")
ax1.imshow(test_im, cmap = 'gray')

ax2.set_title("hue")
ax2.imshow(h, cmap = 'gray')

ax3.set_title('saturation')
ax3.imshow(s, cmap = 'gray')

ax4.set_title('value')
ax4.imshow(v, cmap = 'gray')
```

```python
def avg_brightness(rgb_image):
    hsv = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)
    sum_brightness = np.sum(hsv[:,:,2])
    
    area = 600 *1100
    avg = sum_brightness / area
    
    return avg
```

```python
image_num = 190
test_im = STANDALIZED_LIST[image_num][0]

avg = avg_brightness(test_im)
print("Avg brightness: " +str(avg))
plt.imshow(test_im)
```
