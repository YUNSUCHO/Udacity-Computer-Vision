```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

#read image
image = cv2.imread('images/brain_MR.jpg')
#change colot to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap = 'gray')
```

```python
#try canny using 'tight', 'wide' threshold
wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)

#display the image
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title = ('wide')
ax1.imshow(wide, cmap = 'gray')

ax2.set_title = ('tight')
ax2.imshow(tight, cmap = 'gray')
```

```python
#read the image
image = cv2.imread('images/sunflower.jpg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

lower = cv2.Canny(gray, 30, 100)
upper = cv2.Canny(gray, 200, 240)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title('lower')
ax1.imshow(lower, cmap = 'gray')

ax2.set_title('upper')
ax2.imshow(upper, cmap = 'gray')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

lower = 0
upper = 50

edge = cv2.Canny(gray, lower, upper)

plt.figure(figsize = (20, 10))
plt.imshow(edge, cmap = 'gray')
```
