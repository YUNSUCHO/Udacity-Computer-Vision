```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/water_balloons.jpg')

#make the copy of the image 
image_copy = np.copy(image)

#change the color to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
#plot the color channel
#RGB color channel to isolate
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

#plot
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20, 10))

ax1.set_title("Red")
ax1.imshow(r, cmap = 'gray')

ax2.set_title('Green')
ax2.imshow(g, cmap = 'gray')

ax2.set_title('Blue')
ax3.imshow(b, cmap = 'gray')
```

```python
#convert from RGB to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

h = image[:,:,0]
s = image[:,:,1]
v = image[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20, 10))

ax1.set_title("Hue")
ax1.imshow(h, cmap = 'gray')

ax2.set_title("saturation")
ax2.imshow(s, cmap = 'gray')

ax3.set_title("value")
ax3.imshow(v, cmap = 'gray')
```

```python
#define pink and hue selection threshold
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
```

```python
lower_pink = np.array([180, 0, 100])
upper_pink = np.array([255, 255, 230])
```

```python
#mask image
#define the masked area in rgb color space
mask_rgb = cv2.inRange(image, lower_pink, upper_pink)

#mask the image
mask_image = np.copy(image)
mask_image[mask_rgb==0] = [0, 0, 0]

plt.imshow(mask_image)
```

```python
#defin the masked area in hsv color space
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

#mask the image
mask_image = np.copy(image)
mask_image[mask_hsv==0] = [0, 0, 0]

#display
plt.imshow(mask_image)
```
