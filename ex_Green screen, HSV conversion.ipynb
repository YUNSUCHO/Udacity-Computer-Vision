```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline
```

```python
image = mpimg.imread('images/car_green_screen2.jpg')
plt.imshow(image)
```

```python
#RGB thereshold
lower_green = np.array([0, 180,0])
upper_green = np.array([100, 255, 100])

#define the mask area
mask = cv2.inRange(image, lower_green, upper_green)

#mask the image to let the car show through
mask_image = np.copy(image)
mask_image[mask != 0] = [0, 0, 0]

#display
plt.imshow(mask_image)
```

```python
#convert to hsv
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

#HSV channel
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

#visualize the hsv image
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20, 10))

ax1.set_title("Hue")
ax1.imshow(h, cmap = 'gray')

ax2.set_title("saturation")
ax2.imshow(s, cmap = 'gray')

ax3.set_title("value")
ax3.imshow(v, cmap = 'gray')
```

```python
#define hue threshold
lower_hue = np.array([180, 0, 100])
upper_hue = np.array([50, 255, 255])

#define the masked area
image_hsv = cv2.inRange(image, lower_hue, upper_hue)

mask_image = np.copy(image)
mask_image[image_hsv !=0] = [0, 0, 0]

plt.imshow(mask_image)
```
